This jupyter notebook code file analyse the honest score result over all languages.
For each language, we average over all model with the same k-value. Then we plot the honest score
as k varies and have observed some trends

```python
import csv
import matplotlib.pyplot as plt
languages = ['en','es','fr','it','pt','ro','zh']
result_files_scores = ['./experiment_results/'+lang+'_scores.csv' for lang in languages]
gender = ['1_male','1_female','5_male','5_female','15_male','15_female']
gender_male = ['1_male','5_male','15_male']
gender_female = ['1_female','5_female','15_female']
```

```python
#iterate through male and female model individually, summing the honest score with 
#respect to each k value, then take average.
en_dicts = {}
man={}
female={}
with open(result_files_scores[0], newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    len=0
    #for each type of model
    for row in reader:
        len+=1
        for g in gender_male:
            if g not in man.keys():
                man[g] = float(row[g])
            else:
                man[g] += float(row[g])
        for g in gender_female:
            if g not in female.keys():
                female[g] = float(row[g])
            else:
                female[g] += float(row[g])
    for g in gender_female:
        female[g] /= len
    for g in gender_male:
        man[g] /= len
k_res_male = [man[i] for i in gender_male]
k_res_female = [female[i] for i in gender_female]
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.set_xticks([0,1,2])
ax.set_xticklabels(['k=1','k=5','k=15'])
plt.plot( k_res_male, linewidth=2.0)
plt.plot( k_res_female, linewidth=2.0)
ax.set_title('gender honest score with different k in English')
ax.legend(labels=['Men', 'Women'])
plt.show()
```
